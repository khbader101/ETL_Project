```python
import pandas as pd
from sqlalchemy import create_engine
```

<h2>Extract CSVs into DataFrames</h2>

```python
mcd_file = "McDonalds.csv"
mcd_file_df = pd.read_csv(mcd_file)
mcd_file_df.head()
```

```python
cities_file = "500_Cities.csv"
cities_file_df = pd.read_csv(cities_file)
cities_file_df.head()
```

<h2>Transform McDonald's DataFrame</h2>

```python
#Using .rename(columns={}) in order to rename columns
renamed_df = mcd_file_df.rename(columns={"properties.addressLine1":"address", "properties.addressLine3":"city"}).copy()
renamed_df.head()

#Drop unneeded columns
mcd_drop_df = renamed_df.drop(columns=['geometry.coordinates', 'properties.drivethruhours.driveHoursFriday', 'properties.drivethruhours.driveHoursMonday', 'properties.drivethruhours.driveHoursSaturday', 'properties.drivethruhours.driveHoursSunday',
                                  'properties.drivethruhours.driveHoursThursday', 'properties.drivethruhours.driveHoursTuesday','properties.restaurantUrl', 'properties.restauranthours.hoursFriday', 'properties.restauranthours.hoursMonday',
                                  'properties.restauranthours.hoursSaturday', 'properties.restauranthours.hoursSunday', 'properties.restauranthours.hoursThursday', 'properties.restauranthours.hoursTuesday', 'properties.restauranthours.hoursWednesday',
                                  'properties.subDivision', 'properties.telephone', 'properties.drivethruhours.driveHoursWednesday', 'properties.id', 'properties.identifierValue', 'properties.identifiers.gblnumber', 'properties.identifiers.storeIdentifier',
                                  'properties.openstatus', 'properties.postcode', 'properties.addressLine2'])
mcd_drop_df
```

<h2>Transform Cities DataFrame</h2>

```python
#Using .rename(columns={}) in order to rename columns
renamed_df2 = cities_file_df.rename(columns={"CityName":"city", "Data_Value":"bmi",
                                            "PopulationCount":"population"}).copy()

#Drop unneeded columns
cities_drop_df = renamed_df2.drop(columns=['StateAbbr', 'Year','StateDesc', 'GeographicLevel', 'DataSource', 'Category', 'UniqueID', 'Measure','GeoLocation', 'CategoryID', 
                                           'Data_Value_Footnote','Data_Value_Footnote_Symbol', 'CategoryID', 'Short_Question_Text', 'MeasureId', 'Data_Value_Unit',
                                           'DataValueTypeID', 'Data_Value_Type', 'Low_Confidence_Limit', 'High_Confidence_Limit', 'CityFIPS', 'TractFIPS'])

cap_cities= cities_drop_df['city'].str.upper()
cities_drop_df['population'] = cities_drop_df['population'].str.replace(',','').astype(int)

cities_drop_df=cities_drop_df.assign(city=cap_cities)
cities_drop_df=cities_drop_df.groupby('city').agg({'bmi': 'mean', 'population': 'sum'})
cities_drop_df=cities_drop_df.reset_index()
cities_drop_df
```

 <h2>Create database connection</h2>

```python
connection_string = "postgres:upenn@localhost:5432/McDonalds_and_Obesity"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Confirm tables
engine.table_names()
```

 <h2>Load DataFrames into database</h2>

```python
mcd_drop_df.to_sql(name='mcdonalds', con=engine, if_exists='append', index=False)
```

```python
cities_drop_df.to_sql(name='cities', con=engine, if_exists='append', index=False)
```
